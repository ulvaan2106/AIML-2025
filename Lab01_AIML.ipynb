Part 2: Visualize the data sets using matplotlib libraries

```python
#create the Dataset
people= ['ajay','varun','akash','arun','arjun']
age =[23,24,25,26,27]
weight =[70,71,72,73,74]
height=[170,171,172,173,174]
```

```python
#Find the relationship between weight weight to height
#Generate a scatter plot to visualize the relationship
plt.scatter(weight,height)
plt.xlabel('weight')
plt.ylabel('height')
plt.title('weight vs height')
plt.show()
```

```python
#generate the bar chart for people to their weight
plt.figure(figsize=(5,4))
plt.bar(people,weight)
plt.xlabel('People')
plt.ylabel('Weight')
plt.title('people vs weight')
plt.show()
```

```python
#Generate the Histrogram between people weight
plt.figure(figsize=(5,4))
plt.hist(age,alpha=0.75,color='red',bins=3)
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Age vs Height')
plt.show()
```
